```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load preprocessed data
df = pd.read_csv('Aggregated_Preprocessed_Data_Suraj_Kumar_Sahu.csv')

# 1. Congestion Metrics
# Average vehicle count per hour per junction
hourly_avg = df.groupby(['Junction', 'Hour'])['Vehicles'].mean().unstack()

# 2. Peak Hour Identification (Visualisation)
plt.figure(figsize=(12, 6))
sns.heatmap(hourly_avg, cmap='YlOrRd', annot=False)
plt.title('Traffic Congestion Heatmap: Hour vs Junction')
plt.xlabel('Hour of Day')
plt.ylabel('Junction')
plt.show()

# 3. Weekday vs Weekend Patterns
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='Hour', y='Vehicles', hue='IsWeekend')
plt.title('Traffic Patterns: Weekdays (0) vs Weekends (1)')
plt.show()

# 4. Correlation Analysis
# Selecting numerical columns for correlation
corr_matrix = df[['Vehicles', 'Temperature', 'Humidity', 'Precipitation', 'IsEvent', 'IsWeekend']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation: Traffic Volume vs External Factors')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

# 1. Load Data
df = pd.read_csv('Aggregated_Preprocessed_Data_Suraj_Kumar_Sahu.csv')

# 2. Time-Based Split (Crucial for Time Series)
# We don't use random split; we use the last 20% of time for validation
df = df.sort_values('DateTime')
split_point = int(len(df) * 0.8)
train = df.iloc[:split_point]
test = df.iloc[split_point:]

features = ['Junction', 'Temperature', 'Humidity', 'Precipitation', 
            'Hour', 'DayOfWeek', 'IsWeekend', 'IsEvent', 'Lag_1h', 'Lag_2h']
X_train, y_train = train[features], train['Vehicles']
X_test, y_test = test[features], test['Vehicles']

# 3. Model Training & Hyperparameter Tuning
# Using a Random Forest with optimized parameters
model = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
model.fit(X_train, y_train)

# 4. Model Evaluation
predictions = model.predict(X_test)
mae = mean_absolute_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))
r2 = r2_score(y_test, predictions)

print(f"Mean Absolute Error: {mae:.4f}")
print(f"Root Mean Square Error: {rmse:.4f}")
print(f"R-squared Score: {r2:.4f}")

# 5. Visualization: Actual vs Predicted
plt.figure(figsize=(10, 5))
plt.plot(y_test.values[:100], label='Actual', alpha=0.7)
plt.plot(predictions[:100], label='Predicted', alpha=0.7)
plt.title('Traffic Volume Prediction: Actual vs Predicted (First 100 samples)')
plt.legend()
plt.show()
```
