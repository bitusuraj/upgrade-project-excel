```python
import pandas as pd
import numpy as np

# 1. Load the Integrated Data
# Make sure your 'Integrated_Dataset_Suraj_Kumar_Sahu.csv' is in the same folder
df = pd.read_csv('Integrated_Dataset_Suraj_Kumar_Sahu.csv')
df['DateTime'] = pd.to_datetime(df['DateTime'])

# 2. Data Cleaning
df = df.drop_duplicates()
df['Precipitation'] = df['Precipitation'].fillna(0) # Assume 0 if missing

# 3. Feature Engineering
# Time-based features
df['Hour'] = df['DateTime'].dt.hour
df['DayOfWeek'] = df['DateTime'].dt.dayofweek
df['Month'] = df['DateTime'].dt.month
df['IsWeekend'] = (df['DayOfWeek'] >= 5).astype(int)

# Binary indicator for special events
df['IsEvent'] = (df['Event_Type'] != 'None').astype(int)

# Lag Features: Traffic volume from 1 and 2 hours ago
# This helps the model understand "Temporal Dependencies"
df = df.sort_values(['Junction', 'DateTime'])
df['Lag_1h'] = df.groupby('Junction')['Vehicles'].shift(1)
df['Lag_2h'] = df.groupby('Junction')['Vehicles'].shift(2)

# Drop rows where lag created NaN values
df = df.dropna()

# 4. Normalization (Scaling values between 0 and 1)
cols_to_scale = ['Temperature', 'Humidity', 'Vehicles', 'Lag_1h', 'Lag_2h']
for col in cols_to_scale:
    df[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())

# 5. Exporting for submission
output_name = 'Aggregated_Preprocessed_Data_Suraj_Kumar_Sahu.csv'
df.to_csv(output_name, index=False)
print(f"Dataset preprocessed and saved as {output_name}")
```
